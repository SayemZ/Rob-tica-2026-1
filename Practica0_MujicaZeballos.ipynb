## Previo de la Práctica 0

**ROS 2 (Robot Operating System)** es un conjunto de librerías y herramientas que nos ayudan como desarrolladores a contruir software para aplicaciones robóticas complejas. A pesar del nombre, ROS 2 no es un sistema operativo, si no un framework que se ejecuta sobre un sistema operativo como lo puede ser Linux, Windows o macOS. 

Utilizar ROS nos permite tener una forma estandarizada y modular para comunicar los componentes de software de un robot de manera confiable y eficaz. Además, permite que el desarrollador se enfoque en las fucionalidades específicas de sus nodos sin tener que preocuparse por la comunicación.

En su estructura contaos con **Nodos** y **Tópicos**

- **Nodos**: Es el ejecutable principal de un programa de ROS que se encarga de realizar una función específica dentro del sistema. ROS se compone de múltiples nodos que trabajan juntos y se comunican para cumplir con la función esperada del robot. 

- **Tópicos**: Un tópico es el medio de comunicación asícrono que utilizan los nodos para comunicarse entre sí de forma contínua. Trabajan bajo el model de PubSub, donde el publicador es el nodo que envía los datos y el subscriptor es quien los recibe. Para lograr una comuncación efectiva, los datos deben coincidir en tipo de mensaje.

Dentro del entorno, tenemos una estructura de datos y archivos, en este caso llamados directorios para que un pauqte pueda ser reconocido y cond¿struido correctamente 

```python
my_nodes/
├── CMakeLists.txt
├── package.xml
├── src/
│   └── nodo_publicador.py
├── launch/
│   └── demo_launch.py
└── resource/
```

En esta estructura, **my_nodes** es el directorio raíz del paquete;**CMakeLists.txt**, es la configuración de la construcción donde se define cómo se compila el código; **package.xml**, es el que describe el paquete con su nombre, version, dependencias, mantenedor y licencia; **src**, es donde contiene el código fuente de los ejecutables.

ROS 2 maneja y deifne tre tipos principales de interfaces de comunicación que utilizan los ndos par intercambiar datos

- **Tópicos**: Tipo PubSub, se usa para el envío contínuo de datos como la lectura de sensores y son un *message* simple.

- **Servicios**: de tipo petición-respuesta, usado en operaciones que requieren una respuesta inmediata y un resultado y consiste en un par de estructuras

- **Acciones**: Un modelo de Petición-Meta-Resultado con *feedback* y se utilizan para tareas de larga duración que requieren un *feedback* periódico del progrso. Su mensaje contiene tres partes, una *meta*, un *feedback* y un *resultado*

## Desarrollo de la práctica 

El desarrollo de la práctica se divide en dos partes:

2. El alumno modificará los programas para que los Nodos y Tópicos simulen el control de velocidad de un motor.

Esta práctica consiste en desarrollar dos nodos dentro del entorno de Ros 2, donde uno se encargue de publicar valores de la función seno que varía de 1 a -1 en términos de revoluciones por minuto, y el otro nodo, que cumple el rol de subscriptor, que se encargue de hacer la conversión a radianes por segundo.

```python
#Código del nodo publicador de valores de seno entre -1 y 1
import rclpy
from rclpy.node import Node
from math import sin
from std_msgs.msg import Float64

class MyPublish(Node):
    def __init__(self):
        super().__init__('publish_node')
        self.number_ = 0.0
        self.suma = 0.0
        self.aumento_=0.1
        self.publisher_ =self.create_publisher(Float64, 'publish_topic', 10)
        self.get_logger().info('Nodo publicador activo')
        self.create_timer(0.5,self.subcallback)

    def subcallback(self):
        msg = Float64()
        self.suma += self.aumento_
        msg.data = sin(self.suma)
        self.get_logger().info(f"Publicando seno: {msg.data}")
        self.publisher_.publish(msg)
    
def main(args=None):
    rclpy.init(args=args)
    node = MyPublish()
    rclpy.spin(node)
    rclpy.shutdown()

if __name__ == "__main__":
        main()
```

```python
#Código del nodo subscriptor con el factor de conversión a rad/s
import rclpy
from rclpy.node import Node
from math import sin, pi
from std_msgs.msg import Int32, Float64


class SubNode(Node):
  def __init__(self ):
    super().__init__("node_sub")
    self.mensaje_ = 0
    self.subcriber_ = self.create_subscription(Float64,"publish_topic",self.sub_callbck,10)
    self.vel_rad_motor_ = self.create_publisher(Float64,"rad_vel_topic",10)
    self.get_logger().info("Nodo subcriptor-publicador activo")

  def sub_callbck(self,msg):
    self.mensaje_ += msg.data
    self.new_msg_ = Float64()
    self.new_msg_.data = self.mensaje_*pi/30
    self.get_logger().info(
    f"La velocidad en rad/s es {self.new_msg_.data}"
)
    self.vel_rad_motor_.publish(self.new_msg_)

def main(args=None):
  rclpy.init(args=args)
  node = SubNode()
  rclpy.spin(node)
  rclpy.shutdown()

if __name__ == "__main__":
  main()
```
